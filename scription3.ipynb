## Some useful functions

```python
def get_best_score(grid):
    
    best_score = np.sqrt(-grid.best_score_)
    print(best_score)    
    print(grid.best_params_)
    print(grid.best_estimator_)
    
    return best_score

def print_cols_large_corr(df, nr_c, targ) :
    corr = df.corr()
    corr_abs = corr.abs()
    print (corr_abs.nlargest(nr_c, targ)[targ])
    
def plot_corr_matrix(df, nr_c, targ) :
    
    corr = df.corr()
    corr_abs = corr.abs()
    cols = corr_abs.nlargest(nr_c, targ)[targ].index
    cm = np.corrcoef(df[cols].values.T)

    plt.figure(figsize=(nr_c/1.5, nr_c/1.5))
    sns.set(font_scale=1.25)
    sns.heatmap(cm, linewidths=1.5, annot=True, square=True, 
                fmt='.2f', annot_kws={'size': 10}, 
                yticklabels=cols.values, xticklabels=cols.values
               )
    plt.show()
```

- 목표변수가 정규분포를 따르지 않기 때문에 회귀 모델 성능을 저하시킬수도 있음 -> 로그 변환 필요

|   | Skewness(왜도) | Kurtosis(첨도) |
| --- | --- | --- |
| 정의 | 분포의 비대칭성을 측정 | 분포의 뾰족한 정도를 측정 |
| 계산 공식 | `skew(data)` | `kurtosis(data)` |
| 값의 의미 | 0: 대칭분포<br>양수: 오른쪽으로 치우친 분포<br>음수: 왼쪽으로 치우친 분포 | 3: 정규분포와 같은 뾰족함<br>3보다 크면: 더 뾰족함<br>3보다 작으면: 덜 뾰족함 |
| 응용 | 데이터 분포의 형태 파악, 이상치 탐지 등에 이용 | 회귀 분석, 금융 데이터 분석 등에 이용 |

| 알고리즘 | 특징 | 장점 | 단점 |
| --- | --- | --- | --- |
| 선형 회귀 | 연속적인 종속 변수와 하나 이상의 독립 변수 간의 관계 모델링 | 계산이 빠르고 설명하기 쉬움 | 과적합에 취약하고, 비선형 데이터에서는 성능이 떨어짐 |
| 릿지 회귀(L2) | 선형 회귀에 규제를 추가하여 과적합 방지 | 모든 변수를 사용하므로 정보 손실 최소화 | 적절한 규제 하이퍼파라미터를 찾는 것이 어려울 수 있음 |
| 라쏘 회귀(L1) | 릿지 회귀와 유사하지만 L1 규제 사용 | 중요하지 않은 변수의 계수를 0으로 만들어 특성 선택 가능 | 변수가 많을 경우 모델이 sparse 해질 수 있음 |
| 엘라스틱넷 회귀 | L1 규제와 L2 규제를 함께 사용하여 과적합 방지 | L1 규제로 특성 선택, L2 규제로 계수를 줄임 | 하이퍼파라미터 튜닝이 필요함 |
| 확률적 경사 하강법 회귀 | 반복적으로 가중치 업데이트하여 비용 함수 최적화 | 대용량 데이터 학습 가능, 계산 비용이 적음 | 데이터 스케일링, 하이퍼파라미터 설정이 중요 |
| 결정 트리 회귀 | 결정 트리 알고리즘을 사용한 회귀 | 계산이 간단하고, 시각화 가능 | 과적합에 취약하고, 앙상블 모델보다 성능이 낮음 |
| 랜덤 포레스트 회귀 | 결정 트리의 과적합 문제를 해결하기 위한 앙상블 모델 | 과적합 방지, 변수 중요도 추정 가능 | 하이퍼파라미터 튜닝이 필요함 |
| SVR | Support Vector Machine을 사용한 회귀 | 마진을 최대화하여 회귀 성능 향상 | 커널 함수 선택이 중요하며, 하이퍼파라미터 튜닝이 필요함 |

SGDRegressor

Linear model fitted by minimizing a regularized empirical loss with SGD. SGD stands for Stochastic Gradient Descent: the gradient of the loss is estimated each sample at a time and the model is updated along the way with a decreasing strength schedule (aka learning rate). The regularizer is a penalty added to the loss function that shrinks model parameters towards the zero vector using either the squared euclidean norm L2 or the absolute norm L1 or a combination of both (Elastic Net).

```python
from sklearn.linear_model import SGDRegressor

sgd = SGDRegressor()
parameters = {'max_iter' :[10000], 'alpha':[1e-05], 'epsilon':[1e-02], 'fit_intercept' : [True]  }
grid_sgd = GridSearchCV(sgd, parameters, cv=nr_cv, verbose=1, scoring = score_calc)
grid_sgd.fit(X_sc, y_sc)

sc_sgd = get_best_score(grid_sgd)

pred_sgd = grid_sgd.predict(X_test_sc)
```
